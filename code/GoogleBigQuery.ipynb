# Reading data from GDELT GKG

This module uses Google's BigQuery application programming interface (API) to pull press coverage data from the Global Data on Events Language and Tone (GDELT) Global Knowledge Graph Version 2 (GKG V2) database.  Specifically, it pulls specific press coverage metrics of individual subjects including article and word counts and the GDELT CKG values for tone, positive and negative scores, polarity and activity reference density and self reference density. More information on these scores can be found in the GDELT [documentation](http://data.gdeltproject.org/documentation/GDELT-Global_Knowledge_Graph_Codebook-V2.1.pdf)

It stores these data in a DataFrame and outputs them to a comma separated value (CSV) file.  


## Class objects

To retrieve the coverage metrics, this code employs a class object designed to interact the Google BigQuery Python [API](https://cloud.google.com/bigquery/docs/reference/libraries) and to query the GDELT GKG database.  The class object is GKGPersonPressMetrics and it is developed to retrieve and store in memory press coverage metrics for an individual.  It is built using Google tools including the aforementioned API and the BigQuery public datasets which include GDELT [tables](https://www.gdeltproject.org/data.html#googlebigquery). 

   * GKGPersonPressMetrics attributes:
       * table - the GDELT GKG V2 table at Google BigQuery
       * subject - the name of the individual for whom metrics will be retrieved 
       * start_time - the beginning date/time for the search window
       * end_time - the ending date/time for the search window
       * fields - the GKG V2 fields to retrieve with the query
       * query - the SQL query statement 
       * limit - the maximum number of results to return
       * metrics_ - a dictionary of the returned metrics


   * GKGPersonPressMetrics methods:
       * build_sql() - build the SQL statement used in the press metric search
       * collect_metrics() - execute the search and return a dictionary of metrics
       
## Search string

The GKGPersonPressMetrics object uses a base search string which consists of the following criteria:

   * Select the specified columns from the GKG V2 table
   * Where the GKG V2 Persons field is searched for names matching the subject
   * And the dates fall within the specified interval 
   * Limit the returned output to a maximum



## Data

The data delineated below are the returned and calculated press metrics.  

|Column        |Description    |
|-----------------|--------------------|
|Subject  | name of the individual searched in the Persons field|
|start_time| a datetime object of the search window's start|
|end_time|  a datetime object of the search window's end|
|press_count  | the count of articles including the subject's name and within the time interval|
|word_count| the sum of the word counts of returned articles|
|tone_avg| the average of GDELT's tone measure for the returned articles|
|pos_score_avg  |the average of GDELT's positive score for the returned articles|
|neg_score_avg| the average of GDELT's negative score for the returned articles|
|polarity_avg| the average of GDELT's polarity measure for the returned articles|
|act_ref_den_avg  | the average of GDELT's active reference density measure for the returned articles|
|self_ref_den_avg|  the average of GDELT's self reference density measure for the returned articles|

```python
class GKGPersonPressMetrics:
    # Attributes of the data retrieval
    search_table = "`campfireanalysis.rep_primary_press_coverage_2016.t1`"
    search_paramaters = {}
    fields = []
    time_intervals = {}
    query = ''
    limit = None
    metrics_ = []
    
    # Initialization method
    def __init__(self, search_table = '', search_parameters = {}, 
                 fields = [], limit = None, time_intervals = {}):
        # update self variables
        if search_table != '':
            self.search_table = search_table
        if len(search_parameters) != 0:
            self.search_parameters = search_parameters        
        if len(fields) != 0:
            self.fields = fields
        if limit != None:
            self.limit = limit
        if time_intervals != '':
            self.time_intervals = time_intervals    
      
    
    # Build the SQL string
    def build_sql(self):
        # Construct the select statement
        sel_stm = "SELECT DISTINCT " + ", ".join(self.fields) + " FROM " + self.search_table + " "
        # Construct the subject statement
        sub_stm = "WHERE Persons LIKE('%" + self.search_parameters['subject'].lower() + "%') " 
        # Construct the time statement
        time_stm = "AND DATE > " + self.search_parameters['search_times'][0] + \
                   " AND DATE < " + self.search_parameters['search_times'][1] 
        order_stm = " ORDER BY Date"
        # Construct the limit statement
        if self.limit == None:
            limit_stm = ''
        else:
            limit_stm = " LIMIT "+str(self.limit)

        # Construct a query given the attributes of the class       
        self.query = sel_stm + sub_stm + time_stm + order_stm + limit_stm
        return self.query
        
    
    # method to collect data from posts
    def collect_metrics(self):
        
        # Instantiate the object
        client = bigquery.Client()
        
        # Run the query
        query_job = client.query(self.query, location="US")

        # Check for errors
        if query_job.errors != None:
            if query_job.errors[0]['reason'] == 'invalidQuery':
                # Try to build the sql then run the query again
                self.build_sql()               
                query_job = client.query(self.query, location="US")
            if query_job.errors != None:
                print(query_job.errors)
                raise ValueError('Error with query: ' + query_job.errors[0]['reason'])           

        # Wait for job to finish
        query_job.result()

        # Examine each row returned from the query
        metric = {}
        time_keys = []
        metrics = []
        
        # Build dictionaries for each sub interval in this search window
        for sub_interval in self.search_parameters['sub_dates']:
            time_key = sub_interval[1]
            time_keys.append(time_key)
            metric[time_key] = {}
            metric[time_key]['art_count'] = 0
            metric[time_key]['tone_avg'] = 0
            metric[time_key]['pos_score_avg'] = 0
            metric[time_key]['neg_score_avg'] = 0
            metric[time_key]['polarity_avg'] = 0
            metric[time_key]['act_ref_den_avg'] = 0
            metric[time_key]['self_ref_den_avg'] = 0
            metric[time_key]['word_count'] = 0
            metric[time_key]['themes'] = []

        # Row by row retrieve the returned articles
        for row in query_job:
            # Get the interval key for the this time stamp 
            time_key = interval_key(time = row.get('Date'), times = self.time_intervals)
                      
            # Get the tone metrics from the V2Tone column and store in a dictionary
            #  and keep track of the number of articles in this time_key using count
            try:
                tone_metrics = row.get('V2Tone').split(',')
                metric[time_key]['art_count'] += 1
                metric[time_key]['tone_avg'] += float(tone_metrics[0])
                metric[time_key]['pos_score_avg'] += float(tone_metrics[1])
                metric[time_key]['neg_score_avg'] += float(tone_metrics[2])
                metric[time_key]['polarity_avg'] += float(tone_metrics[3])
                metric[time_key]['act_ref_den_avg'] += float(tone_metrics[4])
                metric[time_key]['self_ref_den_avg'] += float(tone_metrics[5])
                metric[time_key]['word_count'] += float(tone_metrics[6])
            except:
                pass
            
            # Get the themes
            try:
                if row.get('V2Themes') == None or len(metric[time_key]['themes']) > 2_000:
                    pass
                else:
                    metric[time_key]['themes'] += extract_themes(row.get('V2Themes'))
            except:
                pass
            
        # for each time_key in time_keys add some more values to the dictionaries
        for time_key in time_keys:
            # Add the subject to the dictionary
            metric[time_key]['subject'] = self.search_parameters['subject']

            # Store the average tonal scores
            if metric[time_key]['art_count'] != 0:
                metric[time_key]['tone_avg'] = metric[time_key]['tone_avg']/metric[time_key]['art_count']
                metric[time_key]['pos_score_avg'] = metric[time_key]['pos_score_avg']/metric[time_key]['art_count']
                metric[time_key]['neg_score_avg'] = metric[time_key]['neg_score_avg']/metric[time_key]['art_count']
                metric[time_key]['polarity_avg'] = metric[time_key]['polarity_avg']/metric[time_key]['art_count']
                metric[time_key]['act_ref_den_avg'] = metric[time_key]['act_ref_den_avg']/metric[time_key]['art_count']
                metric[time_key]['self_ref_den_avg'] = metric[time_key]['self_ref_den_avg']/metric[time_key]['art_count']
            else:
                metric[time_key]['tone_avg'] = np.NaN
                metric[time_key]['pos_score_avg'] = np.NaN
                metric[time_key]['neg_score_avg'] = np.NaN
                metric[time_key]['polarity_avg'] = np.NaN
                metric[time_key]['act_ref_den_avg'] = np.NaN
                metric[time_key]['self_ref_den_avg'] = np.NaN
       
        # Return the list of dictionaries
        self.metrics_ = metric
        return metric

    
# Use BigQuery to get press information from the search_table and store it 
#  in the table on the dataset
class PressInventory:   
    # Table to search from
    search_table = ''
    # Dataset and table to write to
    dataset_id = ''
    table_id = ''   
    # List of subjects
    subjects = []
    # Tuple of start and end times
    search_times = ()
    # GDELT fields to retrieve
    fields=[]
    # Maximum number of articles to return
    limit = None
    
    # Initialization method
    def __init__(self, search_table = '', dataset_id = '', table_id = '',
                 subjects = [], search_times = [], fields = [], limit = None):
        
        # update class variables        
        if search_table != '':
            self.search_table = search_table
        if dataset_id != '':
            self.dataset_id = dataset_id
        if table_id != '':
            self.table_id = table_id
        if len(subjects) != 0:
            self.subjects = subjects
        if len(search_times) != 0:
            self.search_times = search_times     
        if len(fields) != 0:
            self.fields = fields
        if limit != None:
            self.limit = limit 
                 
    # Build the SQL string
    def build_sql(self):
        # Construct the select statement
        sel_stm = "SELECT DISTINCT " + ", ".join(self.fields) + " FROM " + self.search_table + " "
        # Construct the subject statement
        sub_stm = ["Persons LIKE('%" + subject.lower() +"%') OR" for subject in self.subjects[:-1]]
        sub_stm = "WHERE " + " ".join(sub_stm) + " Persons LIKE('%" + self.subjects[-1].lower() +"%')"
        # Construct the time statement
        time_stm = " AND DATE > " + self.search_times[0] + " AND DATE < " + self.search_times[1]
        # Construct the limit statement
        if self.limit == None:
            limit_stm = ''
        else:
            limit_stm = " LIMIT "+str(self.limit)
            
        # Construct a query given the attributes of the class       
        self.query = sel_stm + sub_stm + time_stm + limit_stm
        return self.query
    
    
    def collect_coverage(self):      
        
        # Instantiate the object
        client = bigquery.Client()

        # Configure BigQuery
        job_config = bigquery.QueryJobConfig()
        # Set the destination table
        table_ref = client.dataset(self.dataset_id).table(self.table_id)
        job_config.destination = table_ref
        # Set job parameters
        job_config.allow_large_results = True
        job_config.write_disposition = 'WRITE_APPEND'

        # Run the query
        query_job = client.query(self.query, 
                                 location="US",
                                 job_config=job_config)
                
        # Check for errors
        if query_job.errors != None:
            if query_job.errors[0]['reason'] == 'invalidQuery':
                # Try to build the sql then run the query again
                self.build_sql()               
                # Run the query
                query_job = client.query(self.query, 
                                         location="US",
                                         job_config=job_config)
            if query_job.errors != None:
                print(query_job.errors)
                raise ValueError('Error with query: ' + query_job.errors[0]['reason'])           
        
        # Wait for job to finish
        query_job.result()
        print('Query results loaded to table {}'.format(table_ref.path))
                
        return 
```

```python
# Retrieve press coverage metrics for a particular candidate
def build_coverage_dataframe(df_polls = None, subject_mapping = {}, intervals = 1,
                             search_table = '', fields =[], search_limit = None,
                             data_path = ''):
 
    # Establish the preferred order of columns
    cols = ['subject','art_count','word_count','tone_avg','pos_score_avg','neg_score_avg',
            'polarity_avg','act_ref_den_avg','self_ref_den_avg','themes']    
    
    # Use the poll dataframe to create polling intervals to 
    #  get the starting and ending times for each poll indexed by ending date 
    poll_intervals = create_poll_intervals(df = df_polls)
    
    # Create a list of search parameters which
    #  sets the subject of each search to one of the candidates
    #  determines search times by breaking the full polling window into n intervals
    #  if the candidate has no poll data for every date in the search window, no entry is created
    search_params = create_search_parameters(df = df_polls, subjects = subject_mapping, intervals = intervals)
    
    # For each set of search parameters
    df_collection = []
    for search_param in search_params:
        

        # Instantiate an search object for this set of search parameters
        person_metric = GKGPersonPressMetrics(search_table=search_table, 
                                              search_parameters = search_param, fields = fields, 
                                              limit=search_limit, time_intervals = poll_intervals)
        # Build the SQL string for this search
        person_metric.build_sql()
        
        # Return a list of dictionaries for this search
        this_search = person_metric.collect_metrics()
        
        # Convert this to a DataFrame
        this_df  = pd.DataFrame(this_search).T
        
        # rearrange the columns
        this_df = this_df[cols]
        
        # Save the search results to a csv as a backup
        file_name = search_param['subject'] + '_' + search_param['search_times'][0][:8] + '_' + \
                    search_param['search_times'][1][:8] + '.csv'
        this_df.to_csv(os.path.join(data_path,file_name))
        
        # Append this DataFrame to the list of DataFrames
        df_collection.append(this_df)
        
    # Return a single DataFrame
    return pd.concat(df_collection, axis = 0)




# # Create a list of tuples with start and end date
# def create_search_dates(df = None):
#     # Set the first date to the first date in the index - 7 days + 1 minute
#     first_date = datetime.strptime(df.index[0],'%Y-%m-%d')-timedelta(days=7)+timedelta(minutes=1)
#     first_date = first_date.strftime('%Y-%m-%d%H%M%S')
#     # Create a list of tuples with start and end dates for each interval
#     search_dates = [(i.replace('-','')+'000100',j.replace('-','')+'235900') \
#                     for i,j in zip(df.index[:-1],df.index[1:])]
#     # Add the first date into the list
#     search_dates.insert(0,(first_date,df.index[0].replace('-','')+'235900'))
#     return search_dates


def create_search_parameters(df = None, subjects = {}, intervals = 1):
     
    # Set the first date to the first date in the index - 7 days + 1 minute
    start_time = datetime.strptime(df.index[0],'%Y-%m-%d')-timedelta(days=7)+timedelta(minutes=1)
    start_date = start_time.strftime('%Y-%m-%d')
    start_time = start_time.strftime('%Y%m%d%H%M%S')
   
    # Break the survey window interval into intervals
    n_breaks = df.shape[0]//intervals
    int_points = [i*n_breaks for i in range(intervals)]
    if df.shape[0]/intervals != 0:
        int_points.append(df.shape[0] - 1 )
    
    # Create a dictionary of search times indexed by the last date in the window
    times = {df.index[j].replace('-','')+'000000': (df.index[i].replace('-','')+'000100', 
             df.index[j].replace('-','')+'235900') for i,j in zip(int_points[:-1],int_points[1:])}
    times[df.index[0].replace('-','')+'000000'] = (start_time,df.index[0].replace('-','')+'235900')
       
    # Create a dictionary that contains the poll periods within each time
    sub_times = {}
    sub_times[df.index[int_points[0]].replace('-','')+'000000'] = [(start_date, df.index[0])]
    for k in range(len(int_points)-1):
        this_interval = [(df.index[i], df.index[i+1]) for i in range(int_points[k],int_points[k+1])]
        sub_times[df.index[int_points[k+1]].replace('-','')+'000000'] = this_interval
    
    # For each candidate and create a search entry if poll results at beginning and end of the interval
    search_params = []
    for subject in subjects.keys():
        # Get the column for this subject
        col_i = df.columns.get_loc(subject)
        
        # Look through each search window
        for i, idx in enumerate(int_points[:-1]):
            # Check to see if poll results are null for each sub_time in this window
            all_null = True
            for j in range(int_points[i],int_points[i+1]):
                if not pd.isnull(df.iloc[j,col_i]):
                    all_null = False
                    break
            # if poll results in the window create a search entry for this candidate
            if all_null == False:
                search_param = {}
                search_param['subject'] = subjects[subject]
                search_param['search_times'] = times[df.index[int_points[i]].replace('-','')+'000000']
                search_param['sub_dates'] = sub_times[df.index[int_points[i]].replace('-','')+'000000']
                search_params.append(search_param)
    return search_params
        

def create_poll_intervals(df = None):
    
    # Set the first date to the first date in the index - 7 days + 1 minute
    start_time = datetime.strptime(df.index[0],'%Y-%m-%d')-timedelta(days=7)+timedelta(minutes=1)
    start_time = start_time.strftime('%Y%m%d%H%M%S')

    # Create a dictionary that contains the poll periods within the dataframe with keys equal to last date
    times = {df.index[j+1]: (df.index[j].replace('-','')+'000100', 
             df.index[j+1].replace('-','')+'235900') for j in range(df.shape[0]-1)}
    times = {df.index[0] : (start_time,df.index[0].replace('-','')+'235900'), **times}
    return times

def interval_key(time = '', times = {}):
    # Search for the time in the dictionary of times
    for interval in times.keys():
        if int(time) >= int(times[interval][0]) and int(time)<=int(times[interval][1]):
            return interval
    return "not_found"
 
def extract_themes(in_text=''):
    return re.findall(r'[\w\d]*_[\w\d]*',in_text)
```
